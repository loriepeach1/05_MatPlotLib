# Analysis Notes
The data is provided for Jan 1 - May 8, 2018.
##  Data Assumptions
* All cities are within the same time zone and within the same general area.
    Miami, FL, and Burlington, VT, are both in the same time zone but would likely have very different trends for Pyber activity.
* All times are local to the time the ride occurred.   
    Factors which impact time analysis include whether the time is from the transaction or from the system storing the data, offsets to GMT, and possibly daylight savings offests.
*  All values are USD.

## City Type
* Rural areas offered the highest fares, and the lowest number of drivers and total rides.
* Urban areas offered the highest number of drivers and rides, with the lowest fare prices.
* Suburban areas were in between with a mid-range number of drivers, total rides, and fare prices. 
* The percent of drivers and fares is distributed similiar across city types.   The percentage of drivers is not as distributed, where 80% are in urban areas.

## Day of Week 
* Across all city types, 
    Sunday was the highest average fare of 28.14.
    <br>Saturday was the lowest average fare of 25.07
    <br>Monday was the busiest day for number of rides, with a grand total of 363.
    <br>Friday was the least busy, with 308.
    

```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime
```

```python
# Read the City and Ride Data
dfCity = pd.read_csv(cityDataCSV)
dfRide = pd.read_csv(rideDataCSV)
```

```python
#the date is a string.   Convert to a date time.  For practice, create a date and time only column.  (not needed for homework)
dfRide["date"] = pd.to_datetime(dfRide["date"])

dfRide['new_date'] = [d.date() for d in dfRide['date']]
dfRide['new_time'] = [d.time() for d in dfRide['date']]
```

```python
# Merge two dataframes using a full join
#Should only need a left merge.  Chose full in the event there was a city with no driver data.
dfMerge = pd.merge(dfRide, dfCity, how="outer", on="city")
dfMerge.head()
```

#### Part A - Create a Scatter Plot.


```python
# Obtain the x and y coordinates for each of the three city types

urban_cities = dfMerge[dfMerge["type"] == "Urban"]
suburban_cities = dfMerge[dfMerge["type"] == "Suburban"]
rural_cities = dfMerge[dfMerge["type"] == "Rural"]

urban_ride_count = urban_cities.groupby(["city"]).count()["ride_id"]
urban_avg_fare = urban_cities.groupby(["city"]).mean()["fare"]
urban_driver_count = urban_cities.groupby(["city"]).mean()["driver_count"]

suburban_ride_count = suburban_cities.groupby(["city"]).count()["ride_id"]
suburban_avg_fare = suburban_cities.groupby(["city"]).mean()["fare"]
suburban_driver_count = suburban_cities.groupby(["city"]).mean()["driver_count"]

rural_ride_count = rural_cities.groupby(["city"]).count()["ride_id"]
rural_avg_fare = rural_cities.groupby(["city"]).mean()["fare"]
rural_driver_count = rural_cities.groupby(["city"]).mean()["driver_count"]
```

```python
# Build the scatter plots for each city types
plt.scatter(urban_ride_count, 
            urban_avg_fare, 
            s=10*urban_driver_count, c="coral", 
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8, label="Urban")

plt.scatter(suburban_ride_count, 
            suburban_avg_fare, 
            s=10*suburban_driver_count, c="skyblue", 
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8, label="Suburban")

plt.scatter(rural_ride_count, 
            rural_avg_fare, 
            s=10*rural_driver_count, c="gold", 
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8, label="Rural")

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)

# Create a legend
lgnd = plt.legend(fontsize="small", mode="Expanded", 
                  numpoints=1, scatterpoints=1, 
                  loc="best", title="City Types", 
                  labelspacing=0.5)
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]

# Incorporate a text label regarding circle size
plt.text(42, 35, "Note:\nCircle size correlates with driver count per city.")

#
#plt.tight_layout()

# Save Figure
plt.savefig("Fig1.png")
plt.savefig("scatter_RideByCity.png", bbox_inches = "tight")

# Show plot
plt.show()
```

### Day and Time Analysis
####  Determine trend by day of week

```python
dfDOW = dfMerge
dfDOW.head()
```

```python
#first convert the string to a datatype
dfDOW["date"] = pd.to_datetime(dfRide["date"])
```

```python
dfDOW.dtypes
```

```python
#View the date/time as unique data

dfDOW['new_date'] = [d.date() for d in dfDOW['date']]
dfDOW['new_time'] = [d.time() for d in dfDOW['date']]
dfDOW['day_of_week'] = dfDOW['date'].dt.weekday_name
dfDOW["dayOfWeek"] = dfDOW['date'].apply(lambda time: time.dayofweek)  #returns index DOW, Monday = 0
dfDOW['day_of_week'] = dfDOW['date'].dt.weekday_name   #returns English dOW
dfDOW.head()
```

```python
def hr_func(ts):
    return ts.hour

dfDOW['time_hour'] = dfDOW['date'].apply(hr_func)
```

```python
dfDOW.head()
```

```python
# Create a df to analyze total rides and average fares by DOW
dfWeekday = dfDOW.groupby("dayOfWeek").agg({"fare": 'mean', "ride_id":'count'}).rename(columns={'fare': 'averageFare', 'ride_id': 'totalRides'})

dfWeekday.head(8)
```

```python
# quick way to plot the data.   Average fare by day of week
dfDOW.groupby("dayOfWeek").agg({"fare": 'mean'}).plot()

plt.title("Average Fare for Day of Week")
plt.xlabel("Day of Week, (0=Monday)")   #need to figure out how to make these the tick marks
plt.ylabel("Average Fare (in USD)")
```

### Day and Time Analysis
####  Determine trend by hour of the day

```python
#determine a trend for the time of day
dfDOW.head()
```

```python
dfHour = dfDOW.groupby("time_hour").agg({"fare": 'mean', "ride_id":'count'}).rename(columns={'fare': 'averageFare', 'ride_id': 'totalRides'})

dfHour.head(24)
```

```python
dfHour2 = dfHour.reset_index() #remove the grouping to use the hour column in plotting
dfHour2.head(24)
```

```python
# Set x axis to hour of the day
# gca stands for 'get current axis'
ax = plt.gca()

dfHour2.plot(kind='line',x='time_hour',y='averageFare',ax=ax)
#dfHour2.plot(kind='line',x='time_hour',y='totalRides', color='red', ax=ax)
# Set a Title and labels
plt.title("Average Fare for Hour of Day")
plt.xlabel("Hour of Day")
plt.ylabel("Fare (in USD)")
plt.show()
```

```python
dfHour3 = dfDOW
dfHour3.head()

```

```python
#use transform to see values by type.
#first determine the total fare by city type, and add this value to the grid
dfHour3["TTLfareByType"] = dfHour3.groupby('type')["fare"].transform('sum')
dfHour3.head()
```

```python
#second, determine the total rides by city type
dfHour3["TTLridesByType"] = dfHour3.groupby('type')["ride_id"].transform('count')
dfHour3.head()
```

```python
#second, determine the AVERAGE fare by city type
dfHour3["AVGfareByType"] = dfHour3.groupby('type')["fare"].transform('mean')
dfHour3.head()
```

```python
#third, determine the AVERAGE number of rides by city type
dfHour3["AVGrideByType"] = dfHour3.groupby('type')["ride_id"].transform('count')
dfHour3.head()
```

```python
#pull the new values from the table into an easy to read grid
pivot = dfHour3.pivot_table(["TTLfareByType","TTLridesByType", "AVGridesByType"], ["type"])
print(pivot)
```

```python
#save the grid into a new df, I need the city type.
summarydf = pivot.reset_index()
summarydf.head()
```
